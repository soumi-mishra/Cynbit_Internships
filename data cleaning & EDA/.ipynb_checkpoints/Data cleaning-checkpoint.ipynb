Load the Dataset


```python
import pandas as pd

# Load training data
df = pd.read_csv('train.csv')


df.head()
df.info()
df.describe()
```

# Data Cleaning


Handle Missing Values


```python
# Total missing values
missing = df.isnull().sum()
missing[missing > 0].sort_values(ascending=False)

# Strategy: fill or drop depending on context
# Example: Drop columns with > 50% missing
df.drop(['Alley', 'PoolQC', 'Fence', 'MiscFeature'], axis=1, inplace=True,errors='ignore')

# Fill others:
df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].median())
df['GarageType'] = df['GarageType'].fillna('None')
print('remaining missing values:\n')
df.head()
```

 Remove Outliers


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visual check for outliers
sns.scatterplot(x='GrLivArea', y='SalePrice', data=df)

# Remove extreme outliers
df = df[df['GrLivArea'] < 4500]
```

Remove Duplicates


```python
# Count number of duplicate rows
num_duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {num_duplicates}")

# Drop duplicates safely 
df = df.drop_duplicates()
```

 Categorical Encoding


```python
# Label Encoding for ordinal
ordinal_cols = ['ExterQual', 'ExterCond', 'BsmtQual', 'KitchenQual']
ordinal_map = {'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1, None: 0}

for col in ordinal_cols:
    df[col] = df[col].map(ordinal_map)

# One-hot encoding for nominal
df = pd.get_dummies(df, drop_first=True)
df.head()
```

Type Conversion:

```python
# Convert numerical to category if needed
df['MSSubClass'] = df['MSSubClass'].astype('category')
df['MoSold'] = df['MoSold'].astype('category')
```

# Before vs After Visualizations


Before Cleaning:

```python
# Missing values heatmap
sns.heatmap(df.isnull(), cbar=False)

# Skewed data
sns.histplot(df['SalePrice'], kde=True)
```

After Cleaning:

```python
# Check distribution after cleaning
sns.histplot(df['SalePrice'], kde=True)

# Correlation heatmap
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=False, cmap='coolwarm')
```
